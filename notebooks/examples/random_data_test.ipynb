```python
import numpy as np
```

```python
N = 10
true_probs = np.random.dirichlet([5, 2, 2], size=N)
print(true_probs)

predicted_probs = true_probs + np.random.normal(0, 0.05, true_probs.shape)
predicted_probs = np.clip(predicted_probs, 0, None)
predicted_probs = predicted_probs / predicted_probs.sum(axis=1, keepdims=True)

print(predicted_probs)
```
