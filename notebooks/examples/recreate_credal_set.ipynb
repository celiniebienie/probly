```python
from matplotlib import tri
import matplotlib.pyplot as plt
import numpy as np
```

```python
def prob_to_xy(p: np.array) -> np.array:
    pa, pb, pc = p
    a = np.array([0, 0])
    b = np.array([1, 0])
    c = np.array([0.5, np.sqrt(3) / 2])
    return pa * a + pb * b + pc * c


a = np.array([0, 0])
b = np.array([1, 0])
c = np.array([0.5, np.sqrt(3) / 2])

probs = np.array(
    [
        [0.8, 0.1, 0.1],  # Mostly class A
        [0.1, 0.8, 0.1],  # Mostly class B
        [0.1, 0.1, 0.8],  # Mostly class C
        [0.33, 0.33, 0.34],  # Near center
    ],
)

coords = np.array([prob_to_xy(p) for p in probs])

plt.figure(figsize=(8, 8))

triangle_x = [a[0], b[0], c[0], a[0]]
triangle_y = [a[1], b[1], c[1], a[1]]
plt.plot(triangle_x, triangle_y, "k-", linewidth=1.5)

plt.scatter(coords[:, 0], coords[:, 1], c=["red", "blue", "green", "purple"])

plt.text(a[0] - 0.04, a[1] - 0.04, "Class A", fontsize=12)
plt.text(b[0] + 0.02, b[1] - 0.04, "Class B", fontsize=12)
plt.text(c[0] - 0.04, c[1] + 0.02, "Class C", fontsize=12)
```

```python
def prob_to_x(p: np.array) -> list:
    pa, pb, pc = p
    a = np.array([0, 0])
    b = np.array([1, 0])
    c = np.array([0.5, np.sqrt(3) / 2])
    x, y = (pa * a + pb * b + pc * c).tolist()
    return x


def prob_to_y(p: np.array) -> list:
    pa, pb, pc = p
    a = np.array([0, 0])
    b = np.array([1, 0])
    c = np.array([0.5, np.sqrt(3) / 2])
    x, y = (pa * a + pb * b + pc * c).tolist()
    return y


def prob_to_coord(p: np.array) -> np.array:
    pa, pb, pc = p
    a = np.array([0, 0])
    b = np.array([1, 0])
    c = np.array([0.5, np.sqrt(3) / 2])
    return pa * a + pb * b + pc * c


probs = np.array(
    [
        [0.1, 0.1, 0.8],
        [0.6, 0.0, 0.4],
        [0.33, 0.33, 0.34],
        [0.05, 0.3, 0.65],
        [0.35, 0.25, 0.4],
    ],
)

points = np.array([prob_to_coord(p) for p in probs])
x_arr = np.array([prob_to_x(p) for p in probs])
y_arr = np.array([prob_to_y(p) for p in probs])

idx_min = np.argmin(points[:, 1])
min_point = points[idx_min]

point_new = np.delete(points, idx_min, axis=0)
print(point_new)
print(idx_min)
print(min_point)

p1, p2, p3, p4 = point_new
print(p1, p2, p3, p4)
```

```python
def prob_to_coord(p: np.array) -> np.array:
    pa, pb, pc = p
    a = np.array([0, 0])
    b = np.array([1, 0])
    c = np.array([0.5, np.sqrt(3) / 2])
    return pa * a + pb * b + pc * c


tri_x = np.array([0.0, 1.0, 0.5])
tri_y = np.array([0.0, 0.0, np.sqrt(3) / 2])

triangle = tri.Triangulation(tri_x, tri_y)

probs = np.array(
    [
        [0.1, 0.1, 0.8],
        [0.6, 0.0, 0.4],
        [0.33, 0.33, 0.34],
        [0.05, 0.3, 0.65],
        [0.35, 0.25, 0.4],
    ],
)
coords = np.array([prob_to_coord(p) for p in probs])
coords_new = np.delete(coords, -1, axis=0)
print(coords_new)
conv_x = coords_new[:, 0]
conv_y = coords_new[:, 1]
conv_x = np.append(conv_x, conv_x[0])
conv_y = np.append(conv_y, conv_y[0])

print(conv_x, conv_y)

# Plot it
plt.figure(figsize=(5, 5))
plt.triplot(triangle, color="k", linewidth=1.5)
plt.scatter(coords[:, 0], coords[:, 1], c=["red", "blue", "green", "purple", "pink"])
plt.plot(conv_x, conv_y, "k-", linewidth=1.5)
plt.fill(conv_x, conv_y, alpha=0.5)
```
