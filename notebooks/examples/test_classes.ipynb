```python
from matplotlib import tri
import matplotlib.pyplot as plt
import numpy as np
```

```python
probs = np.array(
    [
        [0.1, 0.1, 0.8],
        [0.6, 0.0, 0.4],
        [0.33, 0.33, 0.34],
        [0.05, 0.3, 0.65],
        [0.35, 0.25, 0.4],
    ],
)


def probs_to_coord(p: np.array) -> np.array:
    prob_a, prob_b, prob_c = p
    coord_a = np.array([0, 0])
    coord_b = np.array([1, 0])
    coord_c = np.array([0.5, np.sqrt(3) / 2])
    return prob_a * coord_a + prob_b * coord_b + prob_c * coord_c


coords = np.array([probs_to_coord(p) for p in probs])

tri_x = np.array([0, 1, 0.5])
tri_y = np.array([0, 0, np.sqrt(3) / 2])

triangle = tri.Triangulation(tri_x, tri_y)
conv_x = coords[:, 0]
conv_y = coords[:, 1]

plt.figure(figsize=(5, 5))
plt.triplot(triangle)
plt.scatter(conv_x, conv_y)

plt.text(-0.03, -0.03, "Class A", fontsize=8)
plt.text(0.9, -0.03, "Class B", fontsize=8)
plt.text(0.45, 0.88, "Class C", fontsize=8)
```

```python
probs = np.array(
    [
        [0.3, 0.7],
        [0.5, 0.5],
        [0.45, 0.55],
        [0.27, 0.73],
    ],
)


def probs_to_coord(p: np.array) -> np.array:
    prob_a, prob_b = p
    coord_a = np.array([0, 0])
    coord_b = np.array([1, 0])
    return prob_a * coord_a + prob_b * coord_b


coords = np.array([probs_to_coord(p) for p in probs])
print(coords)

x_coords, y_coords = coords[:, 0], coords[:, 1]
x_line = np.array([0, 1])
y_line = np.array([0, 0])

plt.figure(figsize=(6, 3))
plt.plot(x_line, y_line)
plt.scatter(x_coords, y_coords)


epsilon = 0.1  # height of the band
min_x = np.min(x_coords)
max_x = np.max(x_coords)

plt.fill_between([min_x, max_x], -epsilon, epsilon, color="red", alpha=0.2)

plt.xlim(0, 1)
plt.ylim(-1, 1)
```
